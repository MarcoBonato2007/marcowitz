```python
import datetime
import os
import scraper
import json
import pandas as pd
```

```python
NUM_COMPANIES = 2000
```

```python
if not os.path.exists("./data"):
    os.mkdir("./data")
cur_datetime_str = datetime.datetime.now().__str__().replace(" ", "-").replace(":", "-")
rmi = cur_datetime_str.index(".", 1)
cur_datetime_str = cur_datetime_str[:rmi]
```

```python
data_path = f"./data/{cur_datetime_str}"
os.mkdir(data_path)
```

```python
companies = scraper.get_companies()
```

```python
for i in range(NUM_COMPANIES):
    comp = companies[i]
    company_data_path = f"{data_path}/{comp.cik}"
    if (not os.path.exists(company_data_path)):
        os.mkdir(company_data_path)

    # # Get company facts data
    # if not os.path.exists(f"{company_data_path}/companyfacts.json"):
    #     with open(f"{company_data_path}/companyfacts.json", "w") as file:
    #         try:
    #             facts = comp.get_facts()
    #             file.write(json.dumps(facts, indent=4))
    #         except RuntimeError: # some cik's don't have company facts (e.g. s&p 500 trust)
    #             print(f"no comp facts for {comp.name}")
    
    # Get OHCLV data
    if not os.path.exists(f"{company_data_path}/ohclv.csv"):
        with open(f"{company_data_path}/ohclv.csv", "w") as file:
            # https://www.nasdaq.com/market-activity/stocks/{ticker}/historical?page=1&rows_per_page=10&timeline=y10
            # https://finance.yahoo.com/quote/{ticker}/history/?period1=0&period2={really high value}
            try:
                comp.get_ohclv().to_csv(file)
            except:
                print(f"no ohclv for {comp.name}")
```
